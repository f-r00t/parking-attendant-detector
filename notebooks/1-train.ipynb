```python
from ultralytics import YOLO
import os

os.makedirs("../models", exist_ok=True)
model = YOLO("yolov11n.pt")
```

```python
results = model.train(
    data="../data/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    name="yolo_attendant_v1",
    project="../runs"
)
```

```python
import shutil

run_dir = "../runs/detect/yolo_attendant_v1"
best_weights = os.path.join(run_dir, "weights", "best.pt")
dst = "../models/yolov11n_attendant.pt"
shutil.copy(best_weights, dst)
print("Copied best weights to", dst)
```
