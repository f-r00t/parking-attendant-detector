```python
from ultralytics import YOLO

# Load a COCO-pretrained YOLO11n model
model = YOLO("yolo11n.pt")
```

```python
results = model.train(
    data="../data/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    name="yolo_attendant_v1",
    project="../runs"
)
```

```python
import shutil
import os

run_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
best_weights = os.path.join(run_dir, "runs", "yolo_attendant_v13", "weights", "best.pt")
dst = "../models/yolov11n_attendant.pt"
shutil.copy(best_weights, dst)
print("Copied best weights to", dst)
```
